```python
# Importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# Importing data
data = pd.read_csv('Petrol_Gas_Prices_Worldwide.csv', encoding='latin-1')
```

```python
data.head()
```

```python
# We will drop 'S#' column because it will not use in analysis.
data.drop(['S#'], axis=1, inplace =True)
data.head
```

```python
data.shape
```

```python
data.info()
```

**Converting all the object data types to float except country**

```python
data['Daily Oil Consumption (Barrels)']=data['Daily Oil Consumption (Barrels)'].apply(lambda x: x.replace(',',''))
data['Daily Oil Consumption (Barrels)']=data['Daily Oil Consumption (Barrels)'].astype(float)
```

```python
data['GDP Per Capita ( USD )']=data['GDP Per Capita ( USD )'].apply(lambda x: x.replace(',',''))
data['GDP Per Capita ( USD )']=data['GDP Per Capita ( USD )'].astype(float)
```

```python
data['Gallons GDP Per Capita Can Buy']=data['Gallons GDP Per Capita Can Buy'].apply(lambda x: x.replace(',',''))
data['Gallons GDP Per Capita Can Buy']=data['Gallons GDP Per Capita Can Buy'].astype(float)
```

```python
# Converting world share (%age) to floats
Total_Oil_Consumption=data['Daily Oil Consumption (Barrels)'].sum()
data['World Share']=(data['Daily Oil Consumption (Barrels)']/Total_Oil_Consumption)*100
data['World Share']
```

**Performing analysis on top 10 countries**

As in the given dataset, there are many 181 countries. It is difficult to perform analysis on that so, we will perform analysis on top 10 countries.

```python
# Creating a mask of top 10 countries that consume daily oil at maximum amount
Top_10_countries = data.nlargest(10, 'Daily Oil Consumption (Barrels)')
Top_10_countries
```

```python
plt.figure(figsize=(6,3))
Top_contry=Top_10_countries['Country']
Max_oil=Top_10_countries['Daily Oil Consumption (Barrels)']
plt.barh(Top_contry, Max_oil)
plt.title("Daily Oil Consumption by top 10 countries")
plt.xlabel("Daily Oil Consumption (Barrels)")
plt.xlabel("Countries")
plt.tight_layout()
plt.show()
#The graph shows that UNITED STATES has maximum daily oil consumption.
```

```python
plt.figure(figsize=(9,4))
plt.title("Top 10 Countries World Share %age")
ax = sns.barplot(x = "Country", y = "World Share", data = Top_10_countries)
plt.xlabel('Country')
plt.ylabel('World Share')
for i in ax.containers:
    ax.bar_label(i, rotation = 45)  # Shows percentage at each bar.
plt.show()
#The graph shows that UNITED STATES has maximum world share of 20.3851%.
```

```python
data_yearly=Top_10_countries["Yearly Gallons Per Capita"].sort_values(ascending=False)
data_yearly
```

```python
plt.barh(Top_contry,data_yearly)
plt.title("Countries with maximum yearly gallon per capita")
plt.xlabel("Yearly gallon per capita")
plt.tight_layout()
plt.show()
##The graph shows that GERMANY has lowest yearly gallon per capita.
```

```python
GDP=Top_10_countries["GDP Per Capita ( USD )"].sort_values(ascending=False)
GDP
```

```python
plt.barh(Top_contry,GDP )
plt.title("Top 10 Countries vs GDP per capita")
plt.xlabel("GDP per capita")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9,4))
plt.title("Price Per Gallon (USD) of petrol in top 10 countries")
ax = sns.barplot(x = "Country", y = "Price Per Gallon (USD)", data = Top_10_countries)
plt.xlabel('Country')
plt.ylabel('Price Per Liter (pkr)')
for i in ax.containers:
    ax.bar_label(i,)
plt.show()
#The graph shows that SAUDI ARABIA has lowest price per gallon (USD).
```

```python
plt.figure(figsize=(9,4))
plt.title("Price Per Liter (USD) of petrol in top 10 countries")
ax = sns.barplot(x = "Country", y = "Price Per Liter (USD)", data = Top_10_countries)
plt.xlabel('Country')
plt.ylabel('Price Per Liter (USD)')
for i in ax.containers:
    ax.bar_label(i,)
plt.show()
```

```python
plt.figure(figsize=(9,4))
plt.title("Price Per Liter (pkr) of petrol in top 10 countries")
ax = sns.barplot(x = "Country", y = "Price Per Liter (PKR)", data = Top_10_countries)
plt.xlabel('Country')
plt.ylabel('Price Per Liter (pkr)')
for i in ax.containers:
    ax.bar_label(i,)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.scatter("Price Per Liter (USD)","Daily Oil Consumption (Barrels)",sizes=(50, 200),data = Top_10_countries, cmap='Greens', edgecolors='green', linewidths=1, alpha=0.75)
cbar =plt.colorbar()
plt.xlabel('Top 10 countries Daily oil consumption and Price Per Liter (USD)')
plt.tight_layout()
plt.show()
```

Now, we will perform the analysis of pakistan with its neighbouring countries.

```python
# Creating a separate mask of pakistan and its neighbouring countries.
Pak_data=data[data['Country']=='Pakistan']
Ind_data=data[data['Country']=='India']
Iraq_data=data[data['Country']=='Iraq']
Ban_data=data[data['Country']=='Bangladesh']
Afg_data=data[data['Country']=='Afghanistan']
```

```python
# Joining them to form one dataframe.
neighbouring_countries=pd.concat([Pak_data,Ind_data,Iraq_data, Ban_data,Afg_data])
neighbouring_countries
```

```python
sns.distplot(neighbouring_countries['World Share'], rug=True)
```

```python
plt.figure(figsize=(6,3))
plt.title("Neighbouring countries daily oil consumption")
ax = sns.barplot(x = "Country", y = "Daily Oil Consumption (Barrels)", data = neighbouring_countries)
plt.xlabel('Country')
plt.ylabel('Daily Oil Consumption (Barrels)')
for i in ax.containers:
    ax.bar_label(i,)
plt.show()
#The graph shows that INDIA has maximum daily oil consumption.
```

```python
slices = neighbouring_countries['Daily Oil Consumption (Barrels)']
labels = neighbouring_countries['Country']
plt.pie(slices, labels=labels,shadow=True,
        startangle=45, autopct='%1.1f%%',# (shows only 1 decimal point)
        wedgeprops={'edgecolor': 'black'})
plt.title("Daily Oil Consumption (Barrels) by neighbouring countries")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,3))
fig = plt.bar('Country', 'World Share' ,data=neighbouring_countries)
plt.xlabel('Country')
plt.ylabel('World Share')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
#The graph shows that INDIA has maximum world share.
```

```python
slices = neighbouring_countries['World Share']
labels = neighbouring_countries['Country']
plt.figure(figsize=(8,4))
plt.pie(slices, labels=labels,shadow=True,
        startangle=45, autopct='%1.1f%%',# (shows only 1 decimal point)
        wedgeprops={'edgecolor': 'black'})
plt.title("World Share of neighbouring countries")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,3))
fig = plt.bar('Country', 'Yearly Gallons Per Capita' ,data=neighbouring_countries)
plt.xlabel('Country')
plt.ylabel('Yearly Gallons Per Capita')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.title("Neighbouring countries Price Per Liter (USD)")
ax = sns.barplot(x = "Country", y = "Price Per Liter (USD)", data = neighbouring_countries)
plt.xlabel('Country')
plt.ylabel('Price Per Liter (USD)')
for i in ax.containers:
    ax.bar_label(i,)
plt.show()
```
